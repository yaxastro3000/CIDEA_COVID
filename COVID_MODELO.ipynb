## Valle del mezquital.

Las condiciones experimentales mostrarón que solo se presenta el virus en la interfaz aire liquido, por lo que los datos de la interfaz suelo liquido serán cero ($C_{s} =0$, $\mu_{s}= 0$).
El contenido de agua $\theta$ puede interpretarse como la cantidad de liquido en el rio se puede calcular con el flujo del frio de entradas en diversas zonas del rio Tula.

La ecuación diferencial se puede reescribir de la siguiente manera:
    
$$  \frac{\partial^{2} C}{\partial x^{2}}(-\theta D) + \frac{\partial C}{\partial x} (q -\theta \frac{\partial D}{\partial x}- D \frac{\partial \theta}{\partial x})  + C \frac{\partial q}{\partial x}  =  - \mu \theta C - \theta \kappa_{a}^{s} C - \theta \kappa_{a}^{a} C + \theta \kappa_{d}^{a} \theta_{a} C_{a} + \mu_{a}\theta_{a}C_{a}$$

Si consideramos el valor del flujo de Darcy $ q=cte $, es constante: 

$$  \frac{\partial^{2} C}{\partial x^{2}}(-\theta D) + \frac{\partial C}{\partial x} (q -\theta \frac{\partial D}{\partial x}- D \frac{\partial \theta}{\partial x}) =  - \mu \theta C - \theta \kappa_{a}^{s} C - \theta \kappa_{a}^{a} C + \theta \kappa_{d}^{a} \theta_{a} C_{a} + \mu_{a}\theta_{a}C_{a}$$



Para la resolución de la ecuación diferencial, emplearemos una variable $z$ para transformar la ecuación a una de primer orden:

$$ z = \frac{\partial C}{\partial x} $$

Lo que convierte la ecuacion en lo siguiente:

$$  \frac{\partial z}{\partial x}(-\theta D) + z (q -\theta \frac{\partial D}{\partial x}- D \frac{\partial \theta}{\partial x}) =  - \mu \theta C - \theta \kappa_{a}^{s} C - \theta \kappa_{a}^{a} C + \theta \kappa_{d}^{a} \theta_{a} C_{a} + \mu_{a}\theta_{a}C_{a}$$

Siendo estas dos ecuaciones un sistema de ecuaciones diferenciales ordinarias de primer orden para resolver el problema de la dispersión de un virus.

Las condiciones inciales del problea se escribirian como:

$$ z(0) = 0 $$
$$ C(0) = C_{i} $$

$$  \frac{\partial z}{\partial x}(-\theta D) = - z (q -\theta \frac{\partial D}{\partial x}- D \frac{\partial \theta}{\partial x}) +C (- \mu \theta  - \theta \kappa_{a}^{s}  - \theta \kappa_{a}^{a} ) + C_{a}( \theta \kappa_{d}^{a} \theta_{a} + \mu_{a}\theta_{a})$$

```python
# Varibles del problema
q = 1.13 #flujo de Darcy
theta = 1
D = 0.06
mu = 1.3e-5
rhob=19
kas = 8.2e-3
kds = 1.4e-3
kaa = 1.5e-3
kda = 6.3e-3
Ca = 0 #No emplearemos la interfaz agua aire
thetaa = 0 # No es conocida
mua = 0 # se desconoce
K = theta*(-mu-kas-kaa)
R = q - theta*D-D*(0)
#R = q - theta*D*t-D*(0) # dD/dx lo consideramos creciente en distancia.
S = -theta*D
A = Ca*(theta*kda*thetaa + mua*thetaa)
```

```python
#Resolución en Python

from scipy.integrate import odeint
import numpy as np 

def model (x,t):
    y = x[0]
    dy = x[1]
    #K = 30  # -mu theta C ....
    #R = 50  # (q-theta dD/dx - D dtheta/dx)
    #S = 20  # -theta D
    #A = 10  # C_a
    K = theta*(-mu-kas-kaa)
    R = q - theta*D*t-D*(0) # dD/dx lo consideramos creciente en distancia.
    S = -theta*D
    A = Ca*(theta*kda*thetaa + mua*thetaa)
    xdot = [[],[]]
    xdot[0] = dy
    xdot[1] = (R/S)*dy - (K/S)*y + A
    return xdot

time = np.linspace(0,22,1000)
z2 = odeint(model,[1000,0],time)

# plot results
import matplotlib.pyplot as plt
#plt.plot(z1['time'],z1['y'],'r-')
#plt.plot(z1['time'],z1['dy'],'b--')
plt.plot(time,z2[:,0],'g:')
#plt.plot(time,z2[:,1],'k-.')
plt.legend(['y (concentracion)','dy/dx'])
plt.xlabel('Distance')
plt.show()
z2[:,0]
```
